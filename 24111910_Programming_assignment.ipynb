```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Loading the datset and using head function to show first few rows
df=pd.read_excel("/content/Career data_PDA_4053.xlsx")
print("Dataset shape:", df.shape)
print("\n---First 5 rows---")
df.head()
```

```python
# info function gives the data type
print("Information")
df.info()
```

```python
# Using describe function to evaluate the summary
print("\n----Summary Statistics---")
df.describe(include="all")
```

```python
col_Convert=['Age', 'Likely to Change Occupation']

for col in col_Convert:
  if col in col_Convert:
    # Convert to int after filling missing values
    df[col] = pd.to_numeric(df[col], errors='coerce')
    fillup_values= df[col].median() if col=='Age' else df[col].mode()[0]
    df[col]=df[col].fillna(fillup_values)
    df[col] = df[col].round().astype(int)

    print(f"\n{col} column converted to integer successfully.")
else:
    print(f"\nNo {col} column found in dataset.")

print(df.dtypes)
```

```python
# Missing values in dataset
print('Before \nMissing walues:')
df.isnull().sum()
```

```python
# Step 3: Preprocessing

# Handle Missing Values
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include='object').columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
print('After \nMissing Values:')
df.isnull().sum()
```

```python
#Rows count before cleaning rows
print("Rows before:", df.shape[0])
```

```python
# Before vs After Comparison Summary
print("Before Cleaning:")
df[num_cols].describe()
```

```python
# Detect and Treat Outliers (Z-Score method)
df_z = df.copy()
z_scores = np.abs(stats.zscore(df_z[num_cols], nan_policy='omit'))
df_z = df[(z_scores < 3).all(axis=1)]
print("Rows after removing outliers:", df_z.shape[0])
```

```python
# Before vs After Comparison Summary
print("After Cleaning:")
df_z[num_cols].describe()
```

```python
#Step 4: Statistical Checks

# Initial Statistics (Before)
stats_before = df[num_cols].describe()
df[num_cols].describe()
```

```python
# Statistics before cleaning
print("Before:", stats_before)
```

```python
# Statistics After Cleaning
stats_after = df_z[num_cols].describe()
print("Statistical Comparison:\n")
print("After:\n", stats_after)
```

```python
# Years of experience vs frequency (Histogram plot)
plt.figure(figsize=(7,4))
plt.hist(df['Years of Experience'], bins=20,edgecolor='black')
plt.title("Distribution of Work Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Frequency")
plt.show()
```

```python
# Level of Education vs Career change (Bar Chart)
plt.figure(figsize=(9,6))
sns.countplot(data=df, x='Field of Study', hue='Likely to Change Occupation')
plt.title("Level of Education vs Career Move")
plt.xlabel("Field of Study")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Switching Role")
plt.show()
```

```python
# Years of experience vs Likelihood of changing the job (Scatter plot)
plt.figure(figsize=(7,4))
sns.scatterplot(data=df_z, x='Years of Experience', y='Likely to Change Occupation')
plt.title("Experience vs Job Change Likelihood")
plt.xlabel("Years of Experience")
plt.ylabel("Likely to Change Occupation (1 = Yes, 0 = No)")
plt.show()
```

```python
# Boxplot of Age vs Gender(Demographics Graph)
plt.figure(figsize=(6,5))
sns.boxplot(data=df_z, x='Gender', y='Age')
plt.title("Age Distribution Across Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```
